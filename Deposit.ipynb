```python
sex=pd.get_dummies(df['Marital Status'])
sex=sex.iloc[:,:-1]
sex=sex.astype(int)
sex
```

```python
df['Has Credit?'].value_counts()
```

```python
hascredit=pd.get_dummies(df['Has Credit?'],drop_first=True)
hascredit=hascredit.astype(int)
hascredit
```

```python
contact=pd.get_dummies(df['Contact Type'],drop_first=True)
contact=contact.astype(int)
contact
```

```python
df=pd.concat([df,sex,hascredit,contact],axis=1)
df
```

```python
df.drop(['Marital Status','Has Credit?','Contact Type'],inplace=True,axis=1)
```

```python
df
```

```python
df.drop(['Job Status','Education Status','Last Contact Month','Day of the Week'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df
```

```python
df['Subscribed deposit?'].value_counts()
```

```python
df['Subscribed deposit?']=df['Subscribed deposit?'].map({'yes': 1, 'no': 0})
```

```python
df.head()
```

```python
X=df.drop('Subscribed deposit?',axis=1)
y=df['Subscribed deposit?']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lreg=LogisticRegression(max_iter=500)
lreg.fit(X_train,y_train)
```

```python
y_pred=lreg.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print('Classification Report')
print(classification_report(y_test,y_pred))
print('\n')

print('Confusion Matrix')
print(confusion_matrix(y_test,y_pred))
print('\n')


print('Accuracy Score')
print("{} %".format(round(accuracy_score(y_test,y_pred)*100,2)))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(scaled_features,df['TARGET CLASS'],test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
```

```python
print('WITH K=1\n')
print('Classification Report')
print(classification_report(y_test,y_pred_knn))
print('\n')

print('Confusion Matrix')
print(confusion_matrix(y_test,y_pred_knn))
print('\n')


print('Accuracy Score')
print("{} %".format(round(accuracy_score(y_test,y_pred_knn)*100,2)))
```

```python
error_rate=[]

for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn=KNeighborsClassifier(n_neighbors=23)
knn.fit(X_train,y_train)
y_pred_knn2=knn.predict(X_test)
print('WITH K=23\n')
print('Classification Report')
print(classification_report(y_test,y_pred_knn2))
print('\n')

print('Confusion Matrix')
print(confusion_matrix(y_test,y_pred_knn2))
print('\n')


print('Accuracy Score')
print("{} %".format(round(accuracy_score(y_test,y_pred_knn2)*100,2)))

error_rate=[]

for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('K')
plt.ylabel('Error Rate')
```
