```python
import re
def format(time):
    if pd.isna(time):
        return None
    
    match = re.findall(r'\d+', time)
    if not match:
        return None
    
    if 'hrs' in time.lower() or 'hr' in time.lower():
        return int(match[0]) * 60 + (int(match[1]) if len(match) > 1 else 0)
    elif 'days' in time.lower() or 'day' in time.lower():
        return int(match[0]) * 1440 
    return int(match[0])

df['Prep Time (min)']=df['Prep Time'].apply(format)
df['Cook Time (min)']=df['Cook Time'].apply(format)
df
```
